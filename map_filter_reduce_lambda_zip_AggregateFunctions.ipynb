```python
def power_of_5(x):
    return x ** 5

a = [1, 2, 3, 4, 5]
result = map(power_of_5, a)
print(list(result))
```

```python
#filter(function, iterable)
def isOdd(x):
    if x % 2 == 1:
        return True

a = [1, 2, 3, 45, 67, 88, 97, 456]
odd = filter(isOdd, a)
print(odd)
print(list(odd))
```

```python
#reduce()
#functools module
#functools.reduce(function, iterable)
#Calcute the product of a list.
from functools import reduce
def multiplication(x, y):
    return x * y

a = [1, 2, 3, 4, 5, 6, 7, 8, 9]
result = reduce(multiplication, a)
print(result)
```

# lambda function : anonymous function

lambda parameters : expression

**lambda is an expression â€”>**

The value of the expression is returned.

Only an expression can be the body


```python
# Qn . Add 3 numbers

add = lambda x,y,z: x+y+z
```

```python
add(23,56,78)
```

```python
# Check if number is greater than 1,then return number, else return -1

res = lambda a: a if a>10 else -1

res(23)    # inspecting
res(3.5)   # printing

# Note only the last call is in the output ... as return statement is executed only once
# But we can get each called value as output by printing every time we call the lambda function
```

```python
print(res(23))
print(res(3.5))
```

```python
# Return num * 10 if num > 10, return x * 5 if num < 5 otherwise return num

result = lambda num: num * 10 if num > 10 else (num*5 if num < 5 else num)

```

```python
print(result(11))
print(result(2))
print(result(6))
```

# zip function

zip (*iterables) - one or more iterables



image.png

```python
L1 = [1, 2, 3, 4, 5]
L2 = [6, 7, 8, 9, 10]

output = zip(L1,L2)

print(output)
print(list(output))
```

```python
x = [1, 2, 3, 4, 5]
y = ['a', 'b', 'c', 'd', 'e']

result = zip(x,y)

print(list(result))
```

```python
x = [1, 2, 3, 4, 5]

result = zip(x)
print(list(result))
```

```python
m = [1, 3, 50, 23]
n = ['a', 'm']

ans = zip(m,n)

print(ans)

print(list(ans))
```

Q. Multiply corresponding elements of 2 given lists

```python
a = [ 1, 2, 3, 4, 5 ]
b = [ 10, 20, 30, 40, 50 ]

product = [x * y for x,y in zip(a,b)]
print(product)
```
